# Data Augmentation

```python
# example of horizontal flip image augmentation
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import cv2
import os
```

```python
path = 'D:/Documents/Final Year Project/prefinal/Severe'
n = len([name for name in os.listdir(path)])
j = 0
for image_path in os.listdir(path):
    j += 1
    input_path = os.path.join(path, image_path)
    # load the image
    img = load_img(input_path)
    # convert to numpy array
    data = img_to_array(img)
    # expand dimension to one sample
    samples = expand_dims(data, 0)
    # create image data augmentation generator
    datagen = ImageDataGenerator(horizontal_flip=True)
    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    # generate samples
    for i in range(9):
        # generate batch of images
        batch = it.next()
        # convert to unsigned integers for viewing
        print('Creating Image -> '+str(i)+' of -> 9')
        image = batch[0].astype('uint8')
        #writing the images in the directory
        cv2.imwrite('D:/Documents/Final Year Project/gaussian_filtered_images/Severe/fundus_img_severe'+str(j)+str(i)+'.png', image)
        #reading the image for processing
        eye = cv2.imread('D:/Documents/Final Year Project/gaussian_filtered_images/Severe/fundus_img_severe'+str(j)+str(i)+'.png')
        #converting to gray scale
        gray_eye = cv2.cvtColor(eye, cv2.COLOR_BGR2GRAY)
        hist = cv2.equalizeHist(gray_eye)
        # applying CLAHE
        clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8,8))
        cl1 = clahe.apply(gray_eye)
        #gaussian filtering constant
        sigmax = 15
        #applying gaussian filter
        gaussian = cv2.addWeighted(cl1, 4, cv2.GaussianBlur(cl1, (0,0), sigmax), -4, 128)
        final_img = cv2.resize(gaussian, (224, 224), interpolation = cv2.INTER_AREA)
        cv2.imwrite('D:/Documents/Final Year Project/gaussian_filtered_images/Severe/fundus_img_severe'+str(j)+str(i)+'.png', final_img)
```

```python
# example of brighting image augmentation
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
# load the image
img = load_img('dataset/training_set/mild/20051020_43808_0100_PP.png')
# convert to numpy array
data = img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(brightness_range=[0.5,1.0])
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generate samples and plot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```

# Deep Learning Model

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
from keras.layers import BatchNormalization
```

```python
# Initializing the CNN
classifier = Sequential()
```

```python
# Step 1 - CONVOLUTION
classifier.add(Convolution2D(32, (3, 3), input_shape = (64, 64, 3), activation='relu'))
```

```python
# Step 2 - POOLING
classifier.add(MaxPooling2D(pool_size = (2,2)))
```

```python
# Second layer
classifier.add(Convolution2D(32, (3, 3), activation='relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))
# Batch normalization is a technique for training very deep neural networks that standardizes the 
# inputs to a layer for each mini-batch. This has the effect of stabilizing the learning process 
# and dramatically reducing the number of training epochs required to train deep networks.
# classifier.add(BatchNormalization())
```

```python
# Third layer
classifier.add(Convolution2D(64, (3, 3), activation='relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))
# classifier.add(BatchNormalization())
```

```python
# STEP 3 - FLATTENING
classifier.add(Flatten())
```

```python
# STEP 4 - FULL CONNECTION
classifier.add(Dense(activation='relu', units=4096))  #First Hidden layer
classifier.add(Dropout(0.5)) # To prevent overfitting
classifier.add(Dense(activation='relu', units=4096))  #Second Hidden layer

classifier.add(Dense(activation='softmax', units=4)) #Output layer
```

```python
#Compiling the CNN
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#PART 2 - Fitting the CNN to the image
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)

training_set = train_datagen.flow_from_directory('dataset/training_set', 
                                                 target_size = (64,64),
                                                 class_mode = 'categorical')

test_set = test_datagen.flow_from_directory('dataset/test_set',
                                              target_size = (64,64),
                                              class_mode = 'categorical')

early_stopping = EarlyStopping(monitor='val_loss', patience=20, min_delta=0, verbose=1, mode='auto')

hist = classifier.fit_generator(training_set, 
                         samples_per_epoch = 960,
                         epochs = 25,
                         validation_data = test_set,
                         nb_val_samples = 240,
                         shuffle = True,
                         callbacks=[early_stopping])


classifier.evaluate_generator(test_set, 32)
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history["accuracy"])
plt.plot(hist.history['val_accuracy'])
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy","Validation Accuracy","loss","Validation Loss"])
plt.show()
```
